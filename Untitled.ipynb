```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("C:\\Users\\sneha\\Downloads\\SampleSuperstore.csv")
```

```python
df.head(50)
```

```python
# Get information about the dataset
print(df.info())
```

```python
# Check summary statistics of numerical columns
print(df.describe())
```

```python
# Check unique values in categorical columns
print(df['Category'].unique())
print(df['Sub-Category'].unique())
print(df['Segment'].unique())
```

```python
# Check the correlation between numerical columns
correlation = df.corr()
print(correlation)
```

Each value in the matrix represents the correlation between two variables, ranging from -1 to 1. The closer the value is to 1 or -1, the stronger the correlation. A positive value indicates a positive correlation, while a negative value indicates a negative correlation.

For example, we can observe the following correlations:

There is a weak negative correlation (-0.024) between "Postal Code" and "Sales".
There is a moderate positive correlation (0.479) between "Sales" and "Profit".
There is a weak positive correlation (0.201) between "Sales" and "Quantity".
There is a weak negative correlation (-0.219) between "Discount" and "Profit".

```python
# Plot a histogram of the 'Sales' column
plt.hist(df['Sales'], bins=10)
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.title('Distribution of Sales')
plt.show()
```

```python
# Create a bar plot of sales by category
sns.barplot(x='Category', y='Sales', data=df)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
# Create a scatter plot of profit vs. discount
sns.scatterplot(x='Discount', y='Profit', data=df)
plt.xlabel('Discount')
plt.ylabel('Profit')
plt.title('Profit vs. Discount')
plt.show()
```

```python
# Create a box plot of profit by region
sns.boxplot(x='Region', y='Profit', data=df)
plt.xlabel('Region')
plt.ylabel('Profit')
plt.title('Profit by Region')
plt.show()
```

```python
# Create a pair plot of numerical columns
sns.pairplot(df[['Sales', 'Quantity', 'Discount', 'Profit']])
plt.show()
```

```python
# Calculate total sales and profit by category
sales_by_category = df.groupby('Category')['Sales'].sum()
profit_by_category = df.groupby('Category')['Profit'].sum()
```

```python
# Visualize sales and profit by category
plt.figure(figsize=(12, 6))
sns.barplot(x=sales_by_category.index, y=sales_by_category.values)
plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Total Sales by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=profit_by_category.index, y=profit_by_category.values)
plt.xlabel('Category')
plt.ylabel('Profit')
plt.title('Total Profit by Category')
plt.show()
```

```python
# Calculate profit margin by category
df['Profit Margin'] = df['Profit'] / df['Sales']
```

```python
# Visualize profit margin by category
plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Profit Margin', data=df)
plt.xlabel('Category')
plt.ylabel('Profit Margin')
plt.title('Profit Margin by Category')
plt.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import pandas as pd
```

```python
app = dash.Dash(__name__)

app.layout = html.Div(children=[
    html.H1('Business Dashboard'),
# Add a scatter plot
    dcc.Graph(
        id='scatter-plot',
        figure={
            'data': [
                {'x': df['Sales'], 'y': df['Profit'], 'mode': 'markers', 'type': 'scatter'}
            ],
            'layout': {
                'title': 'Sales vs Profit'
            }
        }
    ),
    
    # Add a table
    html.H2('Top 10 Products'),
    html.Table(
        [html.Tr([html.Th(col) for col in df.columns])] +
        [html.Tr([
            html.Td(df.iloc[i][col]) for col in df.columns
        ]) for i in range(10)]
    ),
    
    # Add a dropdown filter
    html.H2('Filter by Category'),
    dcc.Dropdown(
        id='category-dropdown',
        options=[
            {'label': 'Furniture', 'value': 'Furniture'},
            {'label': 'Office Supplies', 'value': 'Office Supplies'},
            {'label': 'Technology', 'value': 'Technology'}
        ],
        value='Furniture'
    ),
    
    # Add a bar chart based on the filter
    dcc.Graph(
        id='category-sales',
        figure={
            'data': [
                {'x': df[df['Category'] == 'Furniture']['Sub-Category'], 'y': df[df['Category'] == 'Furniture']['Sales'], 'type': 'bar', 'name': 'Furniture'},
                {'x': df[df['Category'] == 'Office Supplies']['Sub-Category'], 'y': df[df['Category'] == 'Office Supplies']['Sales'], 'type': 'bar', 'name': 'Office Supplies'},
                {'x': df[df['Category'] == 'Technology']['Sub-Category'], 'y': df[df['Category'] == 'Technology']['Sales'], 'type': 'bar', 'name': 'Technology'}
            ],
            'layout': {
                'title': 'Category-wise Sales'
            }
        }
    )
])
```

```python
# Run your code
app.run_server()
```
